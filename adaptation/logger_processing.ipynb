```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
logger_path = './logger.txt'
with open(logger_path, 'r') as file:
    lines = file.readlines()
    for line in lines:
        print(line.strip())
```

```python
task = []
best_suc = []
demo_num = []
seed = []
task_suite = []
method = []
for i in range(len(lines)):
    line = lines[i].strip()
    if 'evaluate' in line:
        task.append(int(line.split('[')[-1].split(']')[0]))
    if 'best' in line:
        best_suc.append(float(line.split(',')[0].split(':')[-1]))
        task_suite.append(line.split('/')[3])
        seed.append(int(line.split('seed')[-1].split('/')[0].split('_')[-1]))
        demo_num.append(int(line.split('demo')[-1].split('/')[0].split('_')[-1]))
        method.append(line.split('Policy')[0].split('/')[-1])
```

```python
df = pd.DataFrame()
df['task_suite'] = task_suite
df['method'] = method
df['task'] = task
df['seed'] = seed
df['best_suc'] = best_suc
df['demo_num'] = demo_num
```
